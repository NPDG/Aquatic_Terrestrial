## This notebook describes the generation of randomized FAC for each species-source
### These were exported to the same size and overlay them in a different program

```r
library(vegan)
```

```r
brevi_1<-read.csv('b_rand_1.csv')
dim(brevi_1)
```

```r
brevi_1<-brevi_1[,11:648]
brevi_sort_1_fam<-specaccum(brevi_1, method="collector")
```

```r
plot(brevi_sort_1_fam,xlab='Number of isolates',
     ylab='subnetworks',ci.type='poly',col='black',lwd=2,ci.lty=0,
     ylim=c(0,600),
     #xlim=c(0,200),
    ci.col='lightgrey',
    main='P. brevicompactum; random1'
    )
```
