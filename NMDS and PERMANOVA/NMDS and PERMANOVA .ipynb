```r
library(vegan)
```

```r
ex<-read.csv(file="/mnt/c/Users/nawendt/Desktop/GL data/ex_binary3.csv", header=TRUE)
```

```r
df_ex = subset(ex, select = -c(Name, Source) )
row.names(df_ex)<-ex[,1]
```

```r
exDIST <- vegdist(df_ex, method="bray", binary=TRUE, diag=FALSE, upper=FALSE)
```

```r
ex2<-metaMDS(exDIST, distance="bray", autotransform=FALSE)
```

```r
group = c(rep("Group1", 13), rep("Group2", 13))
colors = c(rep("#8B0A07", 13), rep("#FF0000", 13))
ordiplot(ex2)
for(i in unique(group)) {
     ordiellipse(ex2$point[grep(i, group),], draw="polygon", lwd=1,  kind = "se", conf=0.95,
    groups = group[group == i],col = colors[grep(i,group)],label=F, alpha=0.75) 
   ordispider(ex2$point[grep(i, group),], draw="polygon", lwd=2,
    groups = group[group == i],col = colors[grep(i,group)],label=F)}
```

```r
adonis2(exDIST ~ Source, data=ex, method='bray')
```

```r
goodness(ex2)
```

```r
stressplot(ex2)
```
